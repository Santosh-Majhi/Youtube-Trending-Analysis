```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("india_YT_trending.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#check for duplicates
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df["description"] = df["description"].fillna("Not Available")
```

```python
df.info()
```

```python
df["publish_date"] = df["publishedAt"].str.split("T").str[0]
df.head()
```

```python
df["publish_date"] = pd.to_datetime(df["publish_date"])
df.info()
```

```python
gb = df.groupby("channelTitle").agg({"view_count":"sum"})
gb
```

```python
gb = gb.sort_values(by = "view_count", ascending = False)
gb = gb.head()
```

```python
sns.barplot(x = gb.index, y = gb["view_count"], data = gb, hue = gb.index, palette = "viridis")
plt.xticks(rotation = 45)
plt.show()
```

```python
gb1 = df.groupby("title").agg({"view_count":"sum"})
gb1 = gb1.sort_values(by = "view_count", ascending = False)
gb1 = gb1.head()
```

```python
sns.barplot(x = gb1.index, y = gb1["view_count"], data = gb1, hue = gb.index, palette = "spring")
plt.xticks(rotation = 90)
plt.show()
```
